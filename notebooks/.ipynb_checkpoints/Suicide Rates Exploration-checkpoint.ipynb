## Data Preparation and cleaning

```python
# Read the dataset, load into a dataframe, and explore the first 5 rows
df = pd.read_csv('../datasets/suicide_rates.csv')

df.head()
```

### Observations

The dataset is fairly complete, except for the HDI column which is missing the majority of its values.
The gdp_for_year column was inferred as object (string), suggesting some poorly formatting values we might need to clean.

For the purpose of this project, I will attempt to impute the HDI column, which I can do if I have at least 1 value for each country/year pair.

I will also convert and clean the gdp_for_year column, and remove the leading and trailing whitespace from its name.

The age column is an object, because it provides age buckets. While it's useful to bucket like this, it makes it difficult to look at correlation between age and suicide rates for example, so I want to create another column that provides a value based on rank of the age bucket, with 1 being the youngest.

```python
# We will use the re package to build a regex pattern that retains only numerical values
# Then convert the column to an int

def int_conv(value):
    reg_pattern = '[^0-9]'
    new_val = re.sub(reg_pattern, '', value)
    return int(new_val)


df['gdp_for_year_usd'] = df[' gdp_for_year ($) '].apply(int_conv)

df.head()
```

```python
# Now I can drop the first column I don't need anymore (gdp_for_year ($), and give other columns a more coding friendly name)

cols_drop = [' gdp_for_year ($) ']

df = df.drop(cols_drop, axis=1)

col_names = {'suicides/100k pop': 'suicides_per_100k',
            'country-year': 'country_year',
            'gdp_per_capita ($)': 'gdp_per_capita_usd',
            'HDI for year': 'hdi'}

df = df.rename(columns=col_names)

df.head()
```

```python
# Do each unique country_year have an HDI?

df[~df.hdi.isna()].groupby('country_year').count()['hdi']

# No, it looks like we're still missing a lot of data even for a given year. 
# In some cases it looks like we have data every 5 years (Albania for example).
# I'm not planning on using the HDI column in this instance so I'll drop it (and might use external data source later)

df = df.drop(['hdi'], axis=1)
```

```python
# Provide an integer to rank the age buckets where a value of 1 means the youngest age bucket
# Identify the distinct values we can have first, then write a function to apply the correct value.
# We have 6 age categories, so we'll have scores from 1 to 6. 1 being the youngest, 6 the oldest.

map_values = {
    '5-14 years': 1,
    '15-24 years': 2,
    '25-34 years': 3,
    '35-54 years': 4,
    '55-74 years': 5,
    '75+ years': 6
}

df['age_rank'] = df['age'].map(map_values)

# Confirm correct values by finding the min and max age rank and ensure each age group has only one possible value

pd.pivot_table(df, index='age', values='age_rank', aggfunc=['max', 'min'])
```

We now have a clean data frame:
* Columns are all of the right data type for manipulation
* I changed the column names so they're more coding friendly
* I dropped the HDI column, which had a lot of missing values and which I didn't plan on using. 
* I created an age_rank column that helps us get relative "age group" in integer, making correlations feasible.

### Demographics based observations (age, sex, country)

Note: the data is already grouped by age, sex, country, so the data is already normalised to avoid any noise.
In other words, whether a country has a higher rate of females, or an older population, won't affect country based comparisons.

```python
age_pivot = pd.pivot_table(df, index='age', values='suicides_per_100k').sort_values(by='suicides_per_100k')

age_pivot.plot(kind='bar', rot=30, legend=False)

plt.xlabel('Age Group')
plt.ylabel('Suicides per 100K population')
plt.title('''Suicide rates increase relatively linearly with age\nthough we notice that rates increase more rapidly from 5-14 to 15-24 than all other groups''',
          pad=20)

# Set the Y axis to go from 0 to 100
plt.ylim(top=100)

# Add data labels
ys = list(age_pivot['suicides_per_100k'].values)

for i in range(len(ys)):
    plt.text(i - 0.19, ys[i] + 5, str(round(ys[i],1)))

plt.show()
```

```python
sex_pivot = pd.pivot_table(df, index='sex', values='suicides_per_100k').sort_values(by='suicides_per_100k')

sex_pivot.plot(kind='bar', rot=30, legend=False)

plt.xlabel('Sex')
plt.ylabel('Suicides per 100K population')
plt.title('''Men are about 4 times more likely to attempt suicide than women''',
          pad=20)

# Set the Y axis to go from 0 to 100
plt.ylim(top=100)

# Add data labels
ys = list(sex_pivot['suicides_per_100k'].values)

for i in range(len(ys)):
    plt.text(i - 0.05, ys[i] + 5, str(round(ys[i],1)))

plt.show()
```

```python
# There are a lot of countries to show, so I don't want to show all of them on a single chart.
# Instead, I will look at the top 10 countries with highest suicide rates, top 10 with lowest
# And will compare them on a subplot

fig, ax = plt.subplots(nrows=2, ncols=1, sharex=True, figsize=(13,10))

# Pivot table by default calculates the mean, so no need to pass aggfunc
c_pivot = pd.pivot_table(df, index='country', values='suicides_per_100k').sort_values(by='suicides_per_100k', ascending=False)

bottom_10 = c_pivot.tail(10)
top_10 = c_pivot.head(10)

bottom_10.plot(kind='barh', rot=30, ax=ax[0], legend=False, xlabel='')
top_10.plot(kind='barh', rot=30, ax=ax[1], legend=False, xlabel='')

fig.text(0.5, 0.04,'Suicides per 100K')
fig.text(-0.01, 0.5,'Country', rotation='vertical', ha='center', va='center')

# Set the Y axis to go from 0 to 100
plt.xlim([0,100])

ax[0].set_title('Interestingly, we see a lot of islands in low suicide rate countries. These are also all warm countries', pad=20)
ax[1].set_title('With the exception of Sri Lanka, all highest suicide rates countries are in the ex-USSR', pad=20)
plt.show()
```

#### Conclusions
With these three charts we observe that being male and older increases the likelihood to commit suicide.
A number of ex-USSR countries are present in the top suicide countries, which may just be a coincidence.
To better understand why the rates are higher in some countries, we need to look at socio economic backgrounds.

### Socio-Economic profile of countries
In this section we will explore if there are specific features about a country that might explain increased suicide rates.
The information we have is limited: population, and gdp per capita. Therefore conclusions will be limited.

```python
# First, we pivot by country, resetting the index to get the avg suicide rate and gdp across all group in each country year.
# We do this because gdp per capita is given at the country level
economics = pd.pivot_table(data=df, index=['country'], values=['suicides_per_100k', 'gdp_per_capita_usd']).reset_index()
economics['gdp_log'] = np.log(economics['gdp_per_capita_usd'])
# This allows us to build a scatter plot of gdp per capita v suicide rates
sns.scatterplot(data=economics,x='gdp_log', y='suicides_per_100k')

econ_corr = economics.corr()['suicides_per_100k']['gdp_log']

plt.title("There is virtually no correlation between GDP per Capita, and Suicide Rates. R = {}".format(round(econ_corr,2)))
plt.show()
```

```python
# Because there is a large variance in population, I'm going to put it on a log scale so we can see
# all points

pop = pd.pivot_table(data=df, index=['country'], values=['suicides_per_100k', 'population']).reset_index()
pop['population_log'] = np.log(pop['population'])

# This allows us to build a scatter plot of gdp per capita v suicide rates
sns.scatterplot(data=pop,x='population_log', y='suicides_per_100k')

pop_corr = pop.corr()['suicides_per_100k']['population_log']

plt.title("There is a somewhat stronger positive correlation between population, and Suicide Rates. R = {}".format(round(pop_corr,2)))
plt.show()
```

#### Conclusion
Correlation between population, GDP per Capita, and suicide rates is generally very low, even if we see a possible small positive correlation between population and suicide rates.
This suggests there are other factors at play, such as perhaps unemployment, education, weather, political climate etc.

### Trend over the years
In this final part, we want to observe if suicide rates have increased or decreased over the years.
We're going to do it two ways:
1. A line graph showing the overall trend for all countries, and the average across all countries
2. A bar plot showing which countries had the biggest increases or decreases over the years.

```python
df_clean = df.dropna()
year_country = pd.pivot_table(data=df_clean[df_clean.year < 2016], index=['country', 'year'], values='suicides_per_100k').reset_index()
years = pd.pivot_table(data=df_clean[df_clean.year < 2016], index='year', values='suicides_per_100k').reset_index()

# I don't need all countries observations, so to get a clean read I'll drop all NaN valus
year_country = year_country.dropna()

# line plot of all countries, with a low alpha
sns.lineplot(data=year_country, x='year', y='suicides_per_100k', hue='country', alpha=0.1, legend=False)
sns.lineplot(data=years, x='year', y='suicides_per_100k', label='Average Suicide Rates Across the World', lw=2, color='black', linestyle='--')

plt.title("Suicides Rates have been in gradual decline since the mid 90s")
plt.show()
```
